```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#read and display the dataset from the loan.csv excel
ds = pd.read_csv("loan.csv")
ds
```

```python
#Shape of the dataframe no of row and columns in the dataset
ds.shape
```

```python
# data info
ds.info(verbose = True)
```

```python
#summary of ds 
ds.describe(include='O')
```

```python
#Data Cleaning
# check missing value
# find the null values in the column 
ds.isnull().sum()
```

```python
#find the missing value percentage
missing_value_percent = np.round( ds.isnull().sum()/len(ds)*100,2)
missing_value_percent.sort_values(ascending=False)
```

```python
# drop the columns  with 30% missing values not need in the analysis
filterd_ds = ds.loc[:, missing_value_percent <30]
filterd_ds
```

```python
#identify the unique col values
filterd_ds.nunique().sort_values(ascending = True)
```

```python
# identify the coulms with only one unique value
ds_onecolval = filterd_ds.columns[filterd_ds.nunique() ==1]
ds_onecolval
```

```python
# drop coulms with one unique vale not necessaru for analysis
filterd_ds = filterd_ds.drop(ds_onecolval, axis=1)
filterd_ds
```

```python
#find the missing values in filtered ds
filterd_ds.isnull().sum()
```

```python
# fiter the data set by removing all the null value cols
new_filterd_ds = filterd_ds.loc[:,filterd_ds.isnull().sum()==0]
new_filterd_ds.isnull().sum()
```

```python
new_filterd_ds
```

```python
# Data manipulation
# remove % from intrest rate
new_filterd_ds['int_rate'] = new_filterd_ds['int_rate'].str.replace('%','').astype(float)
```

```python
# Remove month from term
new_filterd_ds['term'] = new_filterd_ds['term'].str.replace(" months",'').astype(int)
```

```python
#Create new columns 'year'  and 'month' using issue_d column
new_filterd_ds['issue_d'] = pd.to_datetime(new_filterd_ds['issue_d'], format='%b-%y')
new_filterd_ds['year'] = new_filterd_ds['issue_d'].dt.year
new_filterd_ds['month'] = new_filterd_ds['issue_d'].dt.month
```

```python
new_filterd_ds['year']
```

```python
# Outlier detection using box plot
import matplotlib.pyplot as plt
import seaborn as sns

# loan amount outlier detection
sns.boxplot(new_filterd_ds['loan_amnt'])
plt.title('Loan Amount')
plt.show()
```

```python
# using IQR remove outliers for loan amount

Q1 = new_filterd_ds["loan_amnt"].quantile(0.25)
Q3 = new_filterd_ds["loan_amnt"].quantile(0.75)

IQR = Q3- Q1
IQR
```

```python
# find the upper and lower range 
lower_range = Q1 - 1.5 * IQR
upper_range = Q3 + 1.5 * IQR

lower_range , upper_range
```

```python
#filter the data to remove outlier of loan amonut
data = new_filterd_ds[~((new_filterd_ds['loan_amnt'] < lower_range) | (new_filterd_ds['loan_amnt'] > upper_range))]
data
```

```python
# loan amount after outlier treatment
# loan amount are between 5000 to  15000
sns.boxplot(data['loan_amnt'])
```

```python
# Funded Amount: The total amount committed to that loan at that point in time.

sns.boxplot(data['funded_amnt'])
plt.title('Funded Amount Distribution')
plt.ylabel('Funded Amount')
plt.show()
```

```python
# Remove outlier of funded amount
q3funamnt =  data['funded_amnt'].quantile(0.75) 
q1funamnt = data['funded_amnt'].quantile(0.25)
irq_fnamnt = q3funamnt - q1funamnt
lr_fnamnt = q1funamnt - 1.5 * irq_fnamnt
ur_fnamnt = q3funamnt + 1.5 * irq_fnamnt
#filter the data to remove outlier of funded amount
data = data[~((data['funded_amnt'] < lr_fnamnt) | (new_filterd_ds['funded_amnt'] > ur_fnamnt))]
data
```

```python
# funded amount after outlier treatment
sns.boxplot(data['funded_amnt'])
plt.title('Funded Amount Distribution')
plt.ylabel('Funded Amount')
plt.show()
```
